```python
import numpy as np
import pandas as pd
import seaborn as sns
```

# **Initial Dataset**

```python
df = pd.read_csv("Airbnb Listings Bangkok_raw.csv",
                 parse_dates=['last_review'])
df
```

```python
df.info()
```

# **Data Cleaning**

### **1. Handling Unneeded Columns**

```python
display(df['id'].nunique())
df.shape
```

The column "Unnamed: 0" is unnecessary, since the column "id", which is a unique identifier for each listing, exists.  
We will be dropping this entire column.

```python
df = df.drop(columns='Unnamed: 0')
df
```

### **2. Null Name and Host Name**

```python
df.isna().sum()
```

```python
df[(df['name'].isna()) | (df['host_name'].isna())]
```

The column 'name' and 'host_name' contains null values. We will be deleting all nameless listings, because:
1. The number of listings with no listing name is very insignificant. (9 total vs 15000+ values)
2. The data analysis section may need analysis based on the listing name and host_name. Null values in the dataset may impact the insights obtained from the analysis.
3. An AirBnB listing requires a listing name and a host name (Source: AirBnB Listing Page). Therefore any listing that do not contain these information are likely to be errors in the program, or false information. We will need to remove these to ensure the data we work with are reliable and valid.

```python
df = df.dropna(subset=['name','host_name'])
df
```

### **3. Null Review Values**

```python
df.isna().sum()
```

```python
df[df['last_review'].isna()]
```

Since there are over 5000+ data where the columns 'last_review' is null, we will not be deleting the data.  
We can also notice that most of the columns where 'last_review' is null, the column 'reviews_per_month' is also null.  

```python
df[(df['last_review'].isna()) & (df['number_of_reviews'] == 0)]
```

We can also notice that columns where 'last_reviews' and 'reviews_per_month' is null, the column 'number_of_reviews' is equal to 0.  
Hence these listings contain no reviews at all.  
These will be useful for our data analysis later, so we will be handling these null values.  
<br>
However, we will **not** be handling the column 'last_reviews', because:
1. Attempting to replace the columns with a value, such as 'Unknown', may create inefficiencies during the data analysis section, as the dtype of the column will be converted to 'object'. This will require repeated conversions of the column's dtype back to datetime.
2. There are over 5000+ missing dates on this column, out of 15000+ data, which makes deletion of the data unreasonable, and may lead to biased analysis, in turn leading to incorrect insights.
3. We have analysed above that listings with null 'last_reviews' have exactly zero reviews at all. This can be useful for insight in the data analysis section.

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(value=0)
df
```

### **4. Checking for Duplicates**

```python
df.duplicated().sum()
```

We have confirmed that there are no duplicate rows present in the dataset.

## **5. Price = 0**
An AirBnB listing requires several compulsory information, including price.<br>
Any listing with Price = 0 will be treated as misinput, since AirBnB cannot have a listing with price = 0.

```python
df = df[df['price']!=0]
df
```

The above dataset has now been cleaned. We will be saving this new dataset in a .csv file.
